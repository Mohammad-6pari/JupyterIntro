```python
dfCopy=df.copy(deep=True)
display(dfCopy.info())
mappedValues={
    "divorced":0,
    "single":1,
    "married":2  
}
dfCopy["marital"]=dfCopy["marital"].map(mappedValues)
display(dfCopy.info(),dfCopy.head())
```

## Question3

```python
display("total Nan cells in every column:",df.isna().sum())
```

```python
dfCopy=df.copy(deep=True)
fillMean=lambda col: col.fillna(col.mean()) if (col.dtype != "object") else(col if col.name=="y" else col.fillna(col.mode()[0]))
dfCopy=dfCopy.apply(fillMean, axis=0)
display(dfCopy.info(),dfCopy.head(100))
```

<p dir="rtl" style="text-align:right">خانه های خالی در ستون های عددی را با میانگین اعداد آن ستون و خانه های خالی در ستون های غیرعددی را با mode آن ستون جایگزین کردیم.</p>
<p dir="rtl" style="text-align:right">از مزایای این روش میتوان به راحتی در پیاده سازی، تغییر نکردن میانگین و پراکندگی داده ها قبل و بعد از پرکردن خانه ها و پر نکرن خانه ها با داده های پرت اشاره کرد.</p>

<p dir="rtl" style="text-align:right">از معایب این روش نیز می توان به افزایش تعداد داده های مشابه ،احتمال پر شدن خانه های خالی با مقادیر دور از مقدار واقعی و نامتناسب اشاره کرد</p>

## Question10

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
trainDF=dfCopy[dfCopy["y"].notna()]
indexes=['age','balance','duration','campaign','pdays','y']
numericalDF=trainDF[indexes]
originData=trainDF[indexes]
normalizedDF=(numericalDF-originData.mean())/originData.std()
normalizedDF['y']=trainDF['y']
haveY=normalizedDF[normalizedDF['y']=='yes']
std1,mean1=haveY.std(),haveY.mean()
haveNotY=normalizedDF[normalizedDF['y']=='no']
std2,mean2=haveNotY.std(),haveNotY.mean()
   
display(std1,std2,mean1,mean2)

for col in normalizedDF.columns:
    if col!= "y":
        plt.plot(haveY[col].sort_values(),norm(mean1[col],std1[col]).pdf(haveY[col].sort_values()),label=col+"haveY")
        plt.plot(haveNotY[col].sort_values(),norm(mean2[col],std2[col]).pdf(haveNotY[col].sort_values()),label=col+"notY")
        plt.legend()
        plt.show()
```
